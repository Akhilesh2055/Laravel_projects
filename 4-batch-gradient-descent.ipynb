```python
import numpy as np
import matplotlib.pyplot as plt

# Generate random data-set
X = np.random.rand(100, 1) * 20
Y = np.random.randn(100, 1) + X

print(X.shape)  # (100, 1), 100 rows, 1 column
print(Y.shape)  # (100, 1), 100 rows, 1 column

# Plot the data
plt.scatter(X, Y)
plt.show()
```

```python
mean_x = np.mean(X)
mean_y = np.mean(Y)
n = len(X)

N = 0
D = 0

for i in range(n):
    N += (X[i] - mean_x)*(Y[i] - mean_y)
    D += (X[i] - mean_x) ** 2

b1 = N / D
b0 = mean_y - (b1 * mean_x)

print(b1)
print(b0)
```

```python
# calc b0 and b1 using batch gradient descent

# Initialize parameters
b0 = 0
b1 = 0

# Set hyperparameters
learning_rate = 0.001
iteration = 1000

# Define the number of data points
n = len(X)

# Batch Gradient Descent
for epoch in range(iteration):
    # Compute predictions
    y_pred = b1 * X + b0
    
    # Compute gradients
    db1 = (-2/n) * np.sum(X * (Y - y_pred))
    db0 = (-2/n) * np.sum(Y - y_pred)
    
    # Update parameters
    b1 = b1 - learning_rate * db1
    b0 = b0 - learning_rate * db0

    # print("Epoch: ", epoch, "b1: ", b1, "b0: ", b0)

# Print the final coefficients
print("Coefficients")
print(b1, b0) 
```

```python
x = np.linspace(0, 25, 1000)
y = b1 * x + b0

plt.plot(x,y, color="red")
plt.scatter(X,Y)
plt.show()
```
