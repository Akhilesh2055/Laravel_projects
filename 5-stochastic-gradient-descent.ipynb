```python
import numpy as np
import matplotlib.pyplot as plt

# Generate random data-set
X = np.random.rand(100, 1) * 20
Y = np.random.randn(100, 1) + X

print(X.shape)  # (100, 1), 100 rows, 1 column
print(Y.shape)  # (100, 1), 100 rows, 1 column

# Plot the data
plt.scatter(X, Y)
plt.show()
```

- In SGD, instead of using the entire dataset for each iteration, only a single
  random training example (or a small batch) is selected to calculate the
  gradient and update the model parameters.
- SGD is much faster than GD, but the convergence is not as smooth as GD.
- SGD is more noisy than GD, but it can escape from local minima more easily.
- This random selection introduces randomness into the optimization process,
  hence the term “stochastic” in stochastic Gradient Descent
- The advantage of using SGD is its computational efficiency, especially when dealing with large datasets.


```python
# Initialize parameters
b0 = 0
b1 = 0

# Set hyperparameters
learning_rate = 0.001
iteration = 1000

# Define the number of data points
n = len(X)

# Stochastic Gradient Descent
for epoch in range(iteration):
    for i in range(n):
        # Randomly select a data point
        # Generate a random index between 0 and n-1
        random_index = np.random.randint(n)
        x_i = X[random_index]
        y_i = Y[random_index]
        
        # Compute prediction for the selected point
        y_pred = b1 * x_i + b0
        
        # Compute gradients for the selected point
        db1 = -2 * x_i * (y_i - y_pred)
        db0 = -2 * (y_i - y_pred)
        
        # Update parameters
        b1 = b1 - learning_rate * db1
        b0 = b0 - learning_rate * db0

# Print the final coefficients
print("Coefficients")
print(b1, b0)
```

```python
# Plot the data and regression line
plt.scatter(X, Y)
plt.plot(X, b1*X + b0, color='red')
plt.show()
```
