```python
import numpy as np
import matplotlib.pyplot as plt

# Generate random data-set
X = np.random.rand(100, 1) * 20
Y = np.random.randn(100, 1) + X

print(X.shape)  # (100, 1), 100 rows, 1 column
print(Y.shape)  # (100, 1), 100 rows, 1 column

# Plot the data
plt.scatter(X, Y)
plt.show()
```

- Mini-batch Gradient Descent is a variation of gradient descent that combines
  the benefits of both batch gradient descent and stochastic gradient descent. 
- Instead of using the entire dataset (as in batch GD) or just one data point (as in SGD), mini-batch GD uses a small random subset of the data for each iteration.

```python
# Initialize parameters
b0 = 0
b1 = 0

# Set hyperparameters
learning_rate = 0.001
epochs = 1000
batch_size = 10  # Define the batch size

# Define the number of data points
n = len(X)

# Mini-batch Gradient Descent
for epoch in range(epochs):
    for i in range(0, n, batch_size):
        # Randomly select a mini-batch
        x_batch = X[i:i+batch_size]
        y_batch = Y[i:i+batch_size]
        
        # Compute predictions for the mini-batch
        y_pred = b1 * x_batch + b0
        
        # Compute gradients for the mini-batch
        db1 = (-2/batch_size) * np.sum(x_batch * (y_batch - y_pred))
        db0 = (-2/batch_size) * np.sum(y_batch - y_pred)
        
        # Update parameters
        b1 = b1 - learning_rate * db1
        b0 = b0 - learning_rate * db0

# Print the final coefficients
print("Coefficients")
print(b1, b0)
```

```python
# Plot the data and regression line
plt.scatter(X, Y)
plt.plot(X, b1*X + b0, color='red')
plt.show()
```
