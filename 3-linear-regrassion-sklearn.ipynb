```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Generate random data-set
X = np.random.rand(100, 1) * 20
Y = np.random.randn(100, 1) + X

print(X.shape)  # (100, 1), 100 rows, 1 column
print(Y.shape)  # (100, 1), 100 rows, 1 column

# Plot the data
plt.scatter(X, Y)
plt.show()
```

```python
lin_reg = LinearRegression()
lin_reg.fit(X, Y)

print(lin_reg.coef_, lin_reg.intercept_)
```

```python
b1 = lin_reg.coef_[0][0]
b0 = lin_reg.intercept_[0]

x_new = 22
y_new = b1 * x_new + b0
print(f"Predicated value of {x_new} is {y_new}.") 

x = np.linspace(0, 25, 100)
y = b1 * x + b0

plt.plot(x,y, color="red")
plt.scatter(X,Y)
plt.scatter(x_new,y_new, color="green")
plt.show()
```
