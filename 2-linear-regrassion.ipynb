```python
import numpy as np
import matplotlib.pyplot as plt

# Generate random data-set
X = np.random.rand(100, 1) * 20
Y = np.random.randn(100, 1) + X

print(X.shape)  # (100, 1), 100 rows, 1 column
print(Y.shape)  # (100, 1), 100 rows, 1 column

# Plot the data
plt.scatter(X, Y)
plt.show()
```

```python
mean_x = np.mean(X)
mean_y = np.mean(Y)
n = len(X)

N = 0
D = 0

for i in range(n):
    N += (X[i] - mean_x)*(Y[i] - mean_y)
    D += (X[i] - mean_x) ** 2

b1 = N / D
b0 = mean_y - (b1 * mean_x)

print(b1)
print(b0)
```

```python
x = np.linspace(0, 20, 100)
y = b1 * x + b0

plt.plot(x,y, color="red")
plt.scatter(X,Y)
plt.show()
```

```python
# Predication

x_new = 22
y_new = b1 * x_new + b0
print(f"Predicated value of {x_new} is {y_new}.") 

x = np.linspace(0, 25, 100)
y = b1 * x + b0

plt.plot(x,y, color="red")
plt.scatter(X,Y)
plt.scatter(x_new,y_new, color="green")
plt.show()
```
